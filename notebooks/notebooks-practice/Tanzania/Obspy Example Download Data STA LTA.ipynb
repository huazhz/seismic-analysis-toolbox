```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

client = Client("http://10.254.144.226/")
t = UTCDateTime("2016-06-15T00:00:00.000")
tt = t + 60*24*10
st = client.get_waveforms("TZ", "CES06", "", "HNZ", t, tt)
st.plot()

```

```python
from obspy.signal.trigger import classic_sta_lta
from obspy.signal.trigger import plot_trigger

trace = st[0]
cft = classic_sta_lta(trace.data, int(trace.stats.sampling_rate), int(2.2 * trace.stats.sampling_rate)) # data, small window, large window
plot_trigger(trace, cft, 1.5, 0.5)
```
