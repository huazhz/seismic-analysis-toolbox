```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np

from scipy import signal
from pprint import pprint
```

```python
home = os.getcwd()
eq_path = os.path.join('data', 'eq')
sr_path = os.path.join('data', 'sr')


with open(os.path.join(home, eq_path)) as file:
    eq = [float(line.strip()) for line in file.readlines()]
    
with open(os.path.join(home, sr_path)) as file:
    sr = [float(line.strip()) for line in file.readlines()]
    
def get_range(num, hz):
    """Returns a list of numbers divided by the rate in hz"""
    return np.arange(0, num/hz, 1/hz)  # start, stop, step
    
def unzip(iterable):
    return zip(*iterable)
```

```python
hz = 100  # hz
plt.plot(get_range(len(sr), hz), sr)
plt.xlabel("Time in Seconds")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.figure(figsize=(30,10))

B, A = signal.butter(2, .3, output='ba')
filtered_signal_b2 = signal.filtfilt(B, A, sr)
plt.plot(get_range(len(filtered_signal_b2), hz), filtered_signal_b2)

B, A = signal.butter(8, .3, output='ba')
filtered_signal_b10 = signal.filtfilt(B, A, sr)
plt.plot(get_range(len(filtered_signal_b10), hz), filtered_signal_b10)
plt.show()
```

```python
%matplotlib inline
from functools import wraps
from operator import itemgetter

class Autocorrelation:

    @classmethod
    def auto_correlate(cls, signal, samples, offset, mode='full', method='auto'):
        """
        Computes the autocorrelation of a signal
        
        Params:
            See "partition" method
        
        Returns:
            list(tuple(frame, correlation)): a list containing tuples, each tuple is the frame with it's computed correlation
        """
        frames = cls.partition(signal, samples, offset)
        
        correlations = [] # List of tuples. Each tuple is (frame, correlation).
        
        for frame1 in frames:
            for frame2 in frames:
                correlation = cls.correlate(frame1, frame2, mode=mode, method=method)
                correlations.append((frame1, correlation))
        
        return correlations
    
    @staticmethod
    def partition(signal, samples, offset):
        """
        Splits a signal into "samples" sized frames, separated by offset.
        
        Args:
            param signal: NP array of samples
            param samples: the amount of samples for each frame
            param offset: the separation between each frame (also the separation of overlap)
            
        Returns: 
            list of signals
        """
        index = 0
        frames = []
        # Get the frames 
        for i in range(0, len(signal) - samples, offset):  # start, stop, step
            frame = signal[index:index + samples]
            frames.append(frame)
         
        return frames
     
    @staticmethod
    @wraps(signal.correlate)  # Maintains the original function signature, doc string, and more: https://docs.python.org/2/library/functools.html#functools.wraps
    def correlate(frame1, frame2, *args, **kwargs):   
        """:frame1, frame2, *args, **kwargs: inputs for scipy.signal.correlate"""
        signal.correlate(frame1, frame2, *args, **kwargs)
    
    
ac = Autocorrelation

# frames = ac.partition(sr, 10000, 1000)

correlations = ac.auto_correlate(sr, 10000, 1000)
print(correlations[0])
#correlations = ac.auto_correlate(sr, 1000, 100)
#print(correlations)
```
