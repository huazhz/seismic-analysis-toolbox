```python
from obspy import read
sc = read('https://examples.obspy.org/COP.BHZ.DK.2009.050')
sc.plot(size=(800, 300))
```

```python
dt = sc[0].stats.starttime

sc.plot(color='red', number_of_ticks=7,
                   tick_rotation=5, tick_format='%I:%M %p',
                   starttime=dt + 0, endtime=dt + 60)
```

```python
sc.plot(type='dayplot')
```

```python
import matplotlib.pyplot as plt
from obspy import read

st = read()
tr = st[0]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(tr.times("matplotlib"), tr.data, "g-")
ax.xaxis_date()
fig.autofmt_xdate()
plt.show()
```
