```python
import obspy
from obspy.io.xseed import Parser
from obspy import read, read_inventory
from obspy.clients.fdsn import Client

##########################################
stan="*"   #put 01, 02, *,... for CES01, CES02,...
comp="*"   #put Z, N, E,*
#ti=obspy.UTCDateTime(2016, 7, 13, 12, 0, 0)
#tf=obspy.UTCDateTime(2016, 7, 13, 16, 0, 0)

ti=obspy.UTCDateTime(2016,8, 4,10,24, 4.5)
#ti=obspy.UTCDateTime(2017, 6, 1, 0, 0, 0)
#tf=obspy.UTCDateTime(2016, 10,12,  )
ti= ti -3
tf = ti + 6

#pre_filt = (1,4, 14, 20)
pre_filt = (2,3, 80, 90)
#pre_filt = (10,15, 85, 90)
#pre_filt = (1/80,1/70, 80, 90)

##########################################

##########################################
net='TZ'
staroot="CES"
loc=""
cha="HN"
##########################################


#c = Client("http://jane") 
#c = Client("http://10.254.144.226")
c = Client("http://10.254.145.79")

sta=staroot+stan
chacomp=cha+comp

st = c.get_waveforms(network=net, station='*', location=loc, channel=chacomp,starttime=ti, endtime=tf)
#print(st[0].data)
inv = c.get_stations(network=net, station=sta, location=loc, channel=chacomp, level="response")

stkeep = st.copy()
st.remove_response(inventory=inv, output="VEL",pre_filt=pre_filt,taper=True,water_level=90)
#st[0].plot(type="dayplot",interval=60)
st.plot(equal_scale=False,size=(1000,2000))
filename = str(ti.year)+str(ti.month).zfill(2)+str(ti.day).zfill(2)+ \
        str(ti.hour).zfill(2)+str(ti.minute).zfill(2)+str(ti.second).zfill(2)+'.png'
#st.plot(equal_scale=False,size=(1000,2000),outfile=filename)
```

```python
import matplotlib.pyplot as plt

import numpy as np

tii = ti 
tff = tf 
stsel = st.select(station='CES02')
stsel += st.select(station='CES07')
stsel += st.select(station='CES08')
stsel += st.select(station='CES06')
stsel += st.select(station='CES09')

#stsel.plot(equal_scale=False,size=(1000,2000),starttime=ti,endtime=tf)

#st.select(station='CES06').plot(equal_scale=False,starttime=tii,endtime=tff,size=(600,300))

#a = 9e-9

#tii = ti

#tff = tii +60

#st.select(station='CES09').plot(equal_scale=True,starttime=tii,endtime=tff,size=(600,300)) 
stsel[2].plot(starttime=tii,endtime=tff)

#a = t.axes([0, 8, 1, 100]) print(stsel[2])
for tr in stsel:
    
    tr.plot(starttime=tii,endtime=tff)
    plt.figure(figsize=(8.8,2)) 
    ax = plt.gca()
    tr.spectrogram(per_lap=0.99,wlen=0.3,clip=([0,0.8]),log=True,axes=ax,show=False) 
    plt.axis([0,6,20,90]) 
    plt.ylabel('f [Hz]')
    plt.show()
```
